```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")


file_path = ('C:\\Users\\Admin\\Downloads\\COVID 19 Containment measures data.CSV')
data = pd.read_csv(file_path)


data.info()
data.head()

```

```python
#Distribution of Quantity Variable
plt.figure(figsize=(10, 6))
sns.histplot(data['Quantity'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Distribution of Quantity Variable')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Top 10 Countries by Number of Measures Implemented
plt.figure(figsize=(12, 8))
top_countries = data['Country'].value_counts().head(10) 
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.title('Top 10 Countries by Number of Measures Implemented')
plt.xlabel('Number of Measures')
plt.ylabel('Country')
plt.show()
```

```python
#Correlation Analysis
# Filter numeric columns for correlation matrix
numeric_data = data.select_dtypes(include=['float64', 'int64'])

#  heatmap for correlation if there's more than one numeric column
if numeric_data.shape[1] > 1:
    plt.figure(figsize=(8, 6))
    sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm", square=True)
    plt.title('Correlation Matrix of Numeric Variables')
    plt.show()
else:
    print("Not enough numeric columns for correlation analysis.")
```
